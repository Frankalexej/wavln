```python
import pandas as pd
from paths import *
import os
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
cwlog = pd.read_csv(os.path.join(compound_word_log_path, "log.csv"))
```

```python
cwlog_1 = cwlog[cwlog['match_status'] == 1]
cwlog_0 = cwlog[cwlog['match_status'] == 0]
cwlog_2 = cwlog[cwlog['match_status'] == 2]
```

```python
# Example data (list of numbers)
data = [len(cwlog_1.index), len(cwlog_0.index), len(cwlog_2.index)]

# Labels for each slice of the pie chart
labels = ['Matching', 'Not Matching', 'NAN Included']

# Create a pie chart
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Type of Compounds')  # Set the title
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular

# Show the pie chart
plt.show()
```
